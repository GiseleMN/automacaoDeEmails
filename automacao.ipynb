# Problema
- Passo 1: buscar dados dos valores das ações de forma automática no site do Yahoo Finanças  
- Passo 2: gerar as análises pleo gestor de forma automática 
- Passo 3: enviar um email automaticamente para o gestor com os resultados das análises

## Passo 1

```python
%pip install yfinance
```

```python
import yfinance
```

```python
ticker = input("Digite a sigla da ação que deseja:")
dados = yfinance.Ticker(ticker).history("6mo")
fechamento = dados.Close
fechamento.plot()
```

## Passo 2
#### Análise dos últimos seis meses:
- cotação máxima
- cotação minima
- cotação atual

```python
cot_max = fechamento.max()
cot_min = fechamento.min()
cot_atual = fechamento[-1]
print(cot_max)
print(cot_min)
print(cot_atual)
```

```python
import pyautogui
import pyperclip
```

```python
pyautogui.PAUSE = 2

pyautogui.hotkey("command","t")
pyperclip.copy("www.gmail.com")
pyautogui.hotkey("command","v")
pyautogui.hotkey("enter")
pyautogui.click(x=2033, y=210)

pyperclip.copy("giginisimura@gmail.com")
pyautogui.hotkey("command","v")
pyautogui.hotkey("tab")

pyperclip.copy("analise de dados")
pyautogui.hotkey("command","v")
pyautogui.hotkey("tab")

mensagem = """
Prezado Gestor

Seguem as analises dos ultimos seis meses da ação {ticker}:

Cotação máxima: R$ {round(cot_max,2)}
Cotação minima: R$ {round(cot_min,2)}
cotação Atual: R$ {round(cot_atual,2)}

Qualquer dúvida, estou a disposção

Att.

"""
pyperclip.copy(mensagem)
pyautogui.hotkey("command","v")
pyautogui.click(x=3111,y=971)
```
